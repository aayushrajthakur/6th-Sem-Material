```python
import pandas as pd 
```

```python
file_path = 'dataset/Diwali_Sales_Data.csv'

try:
    diwali_data = pd.read_csv(file_path, encoding='ISO-8859-1')
except UnicodeDecodeError:
    diwali_data = pd.read_csv(file_path, encoding='utf-8', errors='ignore')

print(diwali_data.head())
```

```python
Age_column = 'Age'
```

```python
bins = [0, 18, 35, 50, 100]
```

```python
labels = ['0-18', '19-35', '36-50', '51+']
```

```python
diwali_data['Age Group'] = pd.cut(diwali_data[Age_column], bins=bins, labels=labels, right=False)
```

```python
# Create separate DataFrames for each age group
age_grouped_data = {label: diwali_data[diwali_data['Age Group'] == label] for label in labels}
```

```python
for label, group in age_grouped_data.items():
    print(f'Age Group: {label}')
    print(group.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

plt.figure(figsize=(10, 6)) 
sns.barplot(x='State', y='Product_Category', data=diwali_data)  # Replace 'Category' and 'Value' with your column names

plt.title('Bar Plot of States vs Product_Category')
plt.xlabel('State')
plt.ylabel('Product Category')

# Show the plot
plt.xticks(rotation=45) 
plt.tight_layout() 
plt.show()
```
